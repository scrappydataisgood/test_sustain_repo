```python
import pandas as pd
import scipy as sp
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sn

df = pd.read_csv('/mnt/Datasets/Brazil_CO2_df.csv')
df1 = df.loc[df['result'] == 'emissions_per_area']
df1 = df1.loc[df1['Machine'] == 'herbicide spraying']
df1 = df1.loc[df1['Soil_Drainage'] == 'good']
df1 = df1.loc[df1['Fertilizer_Type'] == 'Compose your own NPK']
df1 = df1.loc[df1['Application_Method'] == 'Incorporate']
yha = df1['Dry_Finished_Yield_Mass_kg']/df1['Crop_Area_ha']

yha = pd.DataFrame(yha)
dfs = pd.concat([yha, df1['Soil_Organic_Matter'], df1['Application_Rate_Unit_kg/ha'], df1['Crop_Protection_Application_Doses'], df1['Total_Water_Applied_mm'], df1['value'], df1['Soil_pH']], axis=1)
dfs = dfs.dropna()

dfs = dfs.rename({0 : 'yield per ha', 'value': 'CO2/ha'},axis=1)
dfs.to_csv('/mnt/out2.csv')

final3 = pd.read_csv('/mnt/out2.csv').drop(columns = ['Unnamed: 0'])

corr_matrix = final3.corr()
plt.figure(figsize = (18, 12))
sn.heatmap(corr_matrix, annot = True)
sn.set(font_scale = 3)
plt.show()
```

```python
df = pd.read_csv('/mnt/Datasets/Brazil_CO2_df.csv')
df1 = df.loc[df['result'] == 'emissions_per_area']
df1 = df1.loc[df1['Machine'] == 'herbicide spraying']
df1 = df1.loc[df1['Soil_Drainage'] == 'good']
df1 = df1.loc[df1['Application_Method'] == 'Incorporate']
yha = df1['Dry_Finished_Yield_Mass_kg']/df1['Crop_Area_ha']

yha = pd.DataFrame(yha)
dfs = pd.concat([yha, df1['value'], df1['Soil_Organic_Matter'], df1['Application_Rate_Unit_kg/ha'], df1['Crop_Protection_Application_Doses'], df1['Total_Water_Applied_mm'], df1['Soil_pH']], axis=1)
dfs = dfs.dropna()

dfs = dfs.rename({0 : 'yield per ha', 'value': 'CO2/ha'},axis=1)
dfs.to_csv('/mnt/out2.csv')

final3 = pd.read_csv('/mnt/out2.csv').drop(columns = ['Unnamed: 0'])

corr_matrix = final3.corr()
plt.figure(figsize = (15, 10))
sn.heatmap(corr_matrix, annot = True)
sn.set(font_scale = 1.5)
plt.show()
print(final3.shape)
```
