```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_1 = pd.read_excel('/mnt/Datasets/PC003_consolidated_new (1).xlsx')
df_1['Field_Name'] = df_1['Contrato legal'].str[6:]
df_1 = df_1.drop(columns = ['Season'])
df_1 = df_1.rename(columns = {'Safra': 'Season'})

print(display(df_1))
```

```python
df_2 = pd.read_csv('/mnt/Datasets/merged_df_1_27_23_w_crop_area.csv').drop(columns = 'Unnamed: 0')
df_2 = df_2.rename(columns = {'Enter Crop Area (ha)': 'Crop_Area_all_subfields', 'Enter Dry Finished Yield Mass (kg)': 'Dry_Yield_Mass_all_subfields'})
df_2['Yield_CO2_ratio'] = df_2['Yield_kg_per_Ha'] / df_2['CO2_kg_per_Ha']

df_2['N_total_kg_per_Ha'] = df_2['N_as_Urea_kg_per_Ha'] + df_2['N_as_Nitrate_kg_per_Ha'] + df_2['N_as_Ammonium_kg_per_Ha']
df_2['P_total_kg_per_Ha'] = df_2['P_as_Phosphorus_kg_per_Ha'] + df_2['P_as_P2O5_kg_per_Ha']
df_2['K_total_kg_per_Ha'] = df_2['K_as_Potassium_kg_per_Ha'] + df_2['K_as_K2O_kg_per_Ha']

df_2 = df_2.drop(columns = ['N_as_Urea_kg_per_Ha', 'N_as_Nitrate_kg_per_Ha', 'N_as_Ammonium_kg_per_Ha', 'P_as_Phosphorus_kg_per_Ha', 'P_as_P2O5_kg_per_Ha', 'K_as_Potassium_kg_per_Ha', 'K_as_K2O_kg_per_Ha'])

print(display(df_2))
```

```python
df = df_2.merge(df_1[['Field_Name', 'Season']], how = 'left', on = 'Field_Name')
df = df.drop_duplicates().reset_index(drop = True)
df = df.drop(columns = ['Total_Water_Applied_mm', 'Crop_Area_all_subfields', 'Dry_Yield_Mass_all_subfields'])

print(display(df))
```

```python
print(df['Season'].value_counts())
```

```python
df_summer = df[df['Season'] == 'SUMMER'].reset_index(drop = True)
df_winter = df[df['Season'] == 'WINTER'].reset_index(drop = True)
```

```python
corr_matrix = df_summer.corr()
plt.figure(figsize = (10, 8))
sns.heatmap(corr_matrix, annot = True, annot_kws={"fontsize":14})
sns.set(font_scale = 1)
plt.show()
```

```python
def rmse(y, y_pred):
    return np.sqrt(np.mean((y - y_pred)**2))

def mae(y, y_pred):
    return np.mean(np.absolute(np.subtract(y, y_pred)))
```
