```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# Load data into a pandas DataFrame
data = pd.read_csv("/mnt/Datasets/df_macrozone_merge_2_3_23.csv").drop(columns = 'Unnamed: 0')

data['Total_N_kg_per_Ha'] = data['N_as_Urea_kg_per_Ha'] + data['N_as_Nitrate_kg_per_Ha'] + data['N_as_Ammonium_kg_per_Ha']
data['Total_P_kg_per_Ha'] = data['P_as_Phosphorus_kg_per_Ha'] + data['P_as_P2O5_kg_per_Ha']
data['Total_K_kg_per_Ha'] = data['K_as_Potassium_kg_per_Ha'] + data['K_as_K2O_kg_per_Ha']

data['Yield_to_CO2_Ratio'] = data['Yield_kg_per_Ha'] / data['CO2_kg_per_Ha']

data = data.drop(columns = ['N_as_Urea_kg_per_Ha', 'N_as_Nitrate_kg_per_Ha', 'N_as_Ammonium_kg_per_Ha', 'P_as_Phosphorus_kg_per_Ha', \
                            'P_as_P2O5_kg_per_Ha', 'K_as_Potassium_kg_per_Ha', 'K_as_K2O_kg_per_Ha', 'CO2_kg_per_Ha', 'Yield_kg_per_Ha', \
                           'Total_Water_Applied_mm'])

p=.95
q_low = data["Yield_to_CO2_Ratio"].quantile(1-p)
q_hi  = data["Yield_to_CO2_Ratio"].quantile(p)

data = data[(data["Yield_to_CO2_Ratio"] < q_hi) & (data["Yield_to_CO2_Ratio"] > q_low)].reset_index(drop = True)

print(display(data))
```
