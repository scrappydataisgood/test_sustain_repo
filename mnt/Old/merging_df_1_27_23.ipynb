```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/mnt/Datasets/Cool-Farm-Tool-GHG-Inputs LATAM_v2.xlsx', sheet_name = 'Fertilizer-Manure')

df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop = True)
df = df.rename(columns = {'Harvest Year': 'Harvest_Year', 'Field Name': 'Field_Name', \
                          'Select Crop Name': 'Crop_Name', 'Enter Crop Area (ha)': 'Crop_Area_Ha', \
                         'Select Fertilizer Type': 'Fertilizer_Type', 'Enter Soil Organic Matter': 'Soil_Organic_Matter', \
                         'Select Manufacturing Region': 'Select_Manufacturing_Region', 'Enter Application Rate': 'Application_Rate_kg_per_Ha', \
                         'Select Rate Unit': 'Rate_Unit', 'Select Units of Product or Nutrient': 'Units_Of_Product_Or_Nutrient', \
                         'Select Application Method': 'Application_Method', 'Select Emissions Inihibitors': 'Emissions_Inhibitors', \
                          '% N as Urea': 'N_as_Urea_%', '% N as Nitrate': 'N_as_Nitrate_%','% N as Ammonium': 'N_as_Ammonium_%', \
                          '% P as Phosphorous': 'P_as_Phosphorus_%', '% P as P2O5': 'P_as_P2O5_%', '% K as Potassium': 'K_as_Potassium_%', \
                          '% K as K2O': 'K_as_K2O_%'})
df = df.drop_duplicates().reset_index(drop = True)
print(display(df))
```

```python
df = df.drop(columns = ['Country', 'Geography', 'Harvest_Year', 'Select_Manufacturing_Region', 'Rate_Unit', \
                        'Units_Of_Product_Or_Nutrient', 'Emissions_Inhibitors'])

print(display(df))
```

```python
df['Application_Rate_kg_per_Ha'] = df['Application_Rate_kg_per_Ha'].astype(float)
df['N_as_Urea_%'] = df['N_as_Urea_%'].astype(float)
df['N_as_Nitrate_%'] = df['N_as_Nitrate_%'].astype(float)
df['N_as_Ammonium_%'] = df['N_as_Ammonium_%'].astype(float)
df['P_as_Phosphorus_%'] = df['P_as_Phosphorus_%'].astype(float)
df['P_as_P2O5_%'] = df['P_as_P2O5_%'].astype(float)
df['K_as_Potassium_%'] = df['K_as_Potassium_%'].astype(float)
df['K_as_K2O_%'] = df['K_as_K2O_%'].astype(float)

print(display(df))
```

```python
df = df.replace('Urea - 46.4% N', 'Urea - 46% N')
df['Fertilizer_Type'].value_counts()
```

```python
#Ammonium Nitrate
df.loc[df['Fertilizer_Type'] == 'Ammonium nitrate - 33.5% N', 'N_as_Nitrate_%'] = 16.9
df.loc[df['Fertilizer_Type'] == 'Ammonium nitrate - 33.5% N', 'N_as_Ammonium_%'] = 16.6

#Ammonium Sulfate
df.loc[df['Fertilizer_Type'] == 'Ammonium sulphate - 21% N', 'N_as_Ammonium_%'] = 21

#Calcium Ammonium Nitrate
df.loc[df['Fertilizer_Type'] == 'Calcium ammonium nitrate - 27% N', 'N_as_Nitrate_%'] = 13.5
df.loc[df['Fertilizer_Type'] == 'Calcium ammonium nitrate - 27% N', 'N_as_Ammonium_%'] = 13.5

#Monammonium Phosphate
df.loc[df['Fertilizer_Type'] == 'Monoammonium phosphate - 11% N / 52% P2O5', 'N_as_Ammonium_%'] = 11
df.loc[df['Fertilizer_Type'] == 'Monoammonium phosphate - 11% N / 52% P2O5', 'P_as_P2O5_%'] = 52

#Potassium chloride
df.loc[df['Fertilizer_Type'] == 'Muriate of potash / Potassium chloride - 60% K2O', 'K_as_K2O_%'] = 60

#Urea 46% N
df.loc[df['Fertilizer_Type'] == 'Urea - 46% N', 'N_as_Urea_%'] = 46

#Urea ammonium nitrate solution
df.loc[df['Fertilizer_Type'] == 'Urea ammonium nitrate solution - 32% N', 'N_as_Urea_%'] = 14.2
df.loc[df['Fertilizer_Type'] == 'Urea ammonium nitrate solution - 32% N', 'N_as_Ammonium_%'] = 8.9
df.loc[df['Fertilizer_Type'] == 'Urea ammonium nitrate solution - 32% N', 'N_as_Nitrate_%'] = 8.9

df = df[(df['Fertilizer_Type'] != 'Compound NPK - 15% N / 15% K2O / 15% P2O5') & (df['Fertilizer_Type'] != 'Poultry layer manure - 1.9% N')].reset_index(drop = True)

df = df.fillna(0)

df.hist(figsize = (10,8))
```

```python
df['N_as_Urea_kg_per_Ha'] = (df['N_as_Urea_%']*df['Application_Rate_kg_per_Ha'])/100
df['N_as_Nitrate_kg_per_Ha'] = (df['N_as_Nitrate_%']*df['Application_Rate_kg_per_Ha'])/100
df['N_as_Ammonium_kg_per_Ha'] = (df['N_as_Ammonium_%']*df['Application_Rate_kg_per_Ha'])/100
df['P_as_Phosphorus_kg_per_Ha'] = (df['P_as_Phosphorus_%']*df['Application_Rate_kg_per_Ha'])/100
df['P_as_P2O5_kg_per_Ha'] = (df['P_as_P2O5_%']*df['Application_Rate_kg_per_Ha'])/100
df['K_as_Potassium_kg_per_Ha'] = (df['K_as_Potassium_%']*df['Application_Rate_kg_per_Ha'])/100
df['K_as_K2O_kg_per_Ha'] = (df['K_as_K2O_%']*df['Application_Rate_kg_per_Ha'])/100
df = df.drop(columns = ['N_as_Urea_%', 'N_as_Nitrate_%', 'N_as_Ammonium_%', 'P_as_Phosphorus_%', 'P_as_P2O5_%', 'K_as_Potassium_%', 'K_as_K2O_%', 'Application_Rate_kg_per_Ha', 'Fertilizer_Type'])

print(display(df))
```
