```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/mnt/Datasets/Cool-Farm-Tool-GHG-Inputs LATAM_v2.xlsx', sheet_name = 'Fertilizer-Manure')

df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop = True)
df = df.rename(columns = {'Harvest Year': 'Harvest_Year', 'Field Name': 'Field_Name', \
                          'Select Crop Name': 'Crop_Name', 'Enter Crop Area (ha)': 'Crop_Area_Ha', \
                         'Select Fertilizer Type': 'Fertilizer_Type', 'Enter Soil Organic Matter': 'Soil_Organic_Matter', \
                         'Select Manufacturing Region': 'Select_Manufacturing_Region', 'Enter Application Rate': 'Application_Rate_kg_per_Ha', \
                         'Select Rate Unit': 'Rate_Unit', 'Select Units of Product or Nutrient': 'Units_Of_Product_Or_Nutrient', \
                         'Select Application Method': 'Application_Method', 'Select Emissions Inihibitors': 'Emissions_Inhibitors', \
                          '% N as Urea': 'N_as_Urea_%', '% N as Nitrate': 'N_as_Nitrate_%','% N as Ammonium': 'N_as_Ammonium_%', \
                          '% P as Phosphorous': 'P_as_Phosphorus_%', '% P as P2O5': 'P_as_P2O5_%', '% K as Potassium': 'K_as_Potassium_%', \
                          '% K as K2O': 'K_as_K2O_%'})

# print(df['Emissions_Inhibitors'].unique())
df = df.drop(columns = ['Country', 'Geography', 'Harvest_Year', 'Select_Manufacturing_Region', 'Rate_Unit', \
                        'Units_Of_Product_Or_Nutrient', 'Emissions_Inhibitors'])

df['Application_Rate_kg_per_Ha'] = df['Application_Rate_kg_per_Ha'].astype(float)
df[['N_as_Urea_%', 'N_as_Nitrate_%','N_as_Ammonium_%', 'P_as_Phosphorus_%','P_as_P2O5_%', 'K_as_Potassium_%', 'K_as_K2O_%']] = \
df[['N_as_Urea_%', 'N_as_Nitrate_%','N_as_Ammonium_%', 'P_as_Phosphorus_%','P_as_P2O5_%', 'K_as_Potassium_%', 'K_as_K2O_%']].fillna(0)

df['N_as_Urea_kg_per_Ha'] = (df['N_as_Urea_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['N_as_Nitrate_kg_per_Ha'] = (df['N_as_Nitrate_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['N_as_Ammonium_kg_per_Ha'] = (df['N_as_Ammonium_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['P_as_Phosphorus_kg_per_Ha'] = (df['P_as_Phosphorus_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['P_as_P2O5_kg_per_Ha'] = (df['P_as_P2O5_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['K_as_Potassium_kg_per_Ha'] = (df['K_as_Potassium_%']*df['Application_Rate_kg_per_Ha']).astype(float)
df['K_as_K2O_kg_per_Ha'] = (df['K_as_K2O_%']*df['Application_Rate_kg_per_Ha']).astype(float)

###################
print(df['Application_Method'].value_counts())
print(df['Fertilizer_Type'].value_counts())

#Consider dropping 'Application_Method' column due to low variance ('Incorporate': 98%; 'Apply in solution': 2%)
#Fertilizer type has some categories that have as low as .05%
###################

df_aggregate_fertilizer_composition = df.groupby('Field_Name')[['N_as_Urea_kg_per_Ha', 'N_as_Nitrate_kg_per_Ha','N_as_Ammonium_kg_per_Ha', \
                                                                'P_as_Phosphorus_kg_per_Ha','P_as_P2O5_kg_per_Ha', 'K_as_Potassium_kg_per_Ha', \
                                                                'K_as_K2O_kg_per_Ha']].sum()

df_aggregate_fertilizer_composition['Field_Name'] = df_aggregate_fertilizer_composition.index
df_aggregate_fertilizer_composition.index.name = ''
df_aggregate_fertilizer_composition = df_aggregate_fertilizer_composition.reset_index(drop = True)

print(display(df_aggregate_fertilizer_composition))
```

```python
df_yield_per_Ha = pd.read_excel('/mnt/Datasets/Cool-Farm-Tool-GHG-Inputs LATAM_v2.xlsx', sheet_name = 'Farm-Crop-Soil')

df_yield_per_Ha.columns = df_yield_per_Ha.iloc[0]
df_yield_per_Ha = df_yield_per_Ha.iloc[1:].reset_index(drop = True)
print(display(df_yield_per_Ha))
df_yield_per_Ha = df_yield_per_Ha.rename(columns = {'Harvest Year': 'Harvest_Year', 'Field Name': 'Field_Name', \
                         'Enter Latitude': 'Latitude', 'Enter Longitude': 'Longitude', \
                         'Select Climate': 'Climate', 'Enter Avg Annual Temp (â„ƒ) - Optional': 'Avg_Annual_Temp_C', \
                          'Select Crop Name': 'Crop_Name', 'Enter Crop Area (ha)': 'Crop_Area_Ha', \
                         'Enter Seed Amount (If Potato)': 'Seed_Amount', 'Select Seed Amount Unit (If Potato)': 'Seed_Amount_Unit', \
                         'Enter Total Wet Mass Harvested (kg)': 'Wet_Mass_Harvested_kg', 'Enter Dry Finished Yield Mass (kg)': 'Dry_Yield_Mass_Kg', \
                         'Enter Residue Amount (kg/ha) - Optional': 'Residue_Amount', 'Selct Residue Management Practice': 'Residue_Management_Practice', \
                         'Select Soil Texture': 'Soil_Texture', 'Enter Soil Organic Matter': 'Soil_Organic_Matter', \
                         'Select Soil Moisture ': 'Soil_Moisture', 'Select Soil Drainage': 'Soil_Drainage', \
                         'Enter Soil pH': 'Soil_pH'})

df_yield_per_Ha['Yield_Mass_kg_Per_Ha'] = df_yield_per_Ha['Dry_Yield_Mass_Kg'] / df_yield_per_Ha['Crop_Area_Ha']

# get rid of Soil_Drainage, Soil_pH due to low variance (filter only soil drainage by = 'good')

df_yield_per_Ha = df_yield_per_Ha[df_yield_per_Ha['Soil_Drainage'] == 'good'].reset_index(drop = True)

df_yield_per_Ha = df_yield_per_Ha.drop(columns = ['Country', 'Geography', 'Climate', 'Crop_Name', 'Seed_Amount', 'Seed_Amount_Unit', \
                        'Wet_Mass_Harvested_kg', 'Latitude', 'Longitude', 'Residue_Amount', 'Soil_Texture', \
                        'Residue_Management_Practice', 'Avg_Annual_Temp_C', 'Soil_Moisture', 'Harvest_Year', 'Dry_Yield_Mass_Kg', 'Soil_Drainage', 'Soil_pH'])

print(display(df_yield_per_Ha))
```

```python
df_merged = df_aggregate_fertilizer_composition.merge(df_yield_per_Ha, on = 'Field_Name', how = 'left')

# fields_in_aggregate = df_aggregate_fertilizer_composition['Field_Name'].unique()
# df_yield_per_Ha = df_yield_per_Ha[df_yield_per_Ha['Field_Name'].isin(fields_in_aggregate)].reset_index(drop = True)

#N_as_Urea_kg_per_Ha, N_as_Ammonium_kg_per_Ha, P_as_P2O5_kg_per_Ha, K_as_K2O_kg_per_Ha (low variance)

df_merged = df_merged.drop(columns = ['Soil_pH', 'Soil_Drainage', 'N_as_Urea_kg_per_Ha', 'N_as_Ammonium_kg_per_Ha', 'P_as_P2O5_kg_per_Ha', 'K_as_K2O_kg_per_Ha'])

print(display(df_merged))

print(display(df_merged[df_merged.isna()]))
```
