```python
import re

#Converting latitude to decimal
temp = df_complete[df_complete['result'] == 'emissions_total'].copy()
temp['Latitude'] = temp['Latitude'].str.replace(',', '.', regex=False)
temp['Latitude'] = temp['Latitude'].str.replace('S ', '', regex=False)
temp['Latitude'] = temp['Latitude'] + ' S'
temp['Latitude'] = temp['Latitude'].str.replace(' ', '', regex=False)

pattern = r'(?P<d>[\d\.]+).*?(?P<m>[\d\.]+).*?(?P<s>[\d\.]+).*?(?P<h>[NSWE])'
hemisphere_sig = {'N': 1, 'S': -1, 'E': 1, 'W': -1}

dms = temp['Latitude'].str.extract(pattern)
dms['h'] = dms['h'].map(hemisphere_sig)
dms = dms.astype(float)
temp['Latitude'] = dms['h'] * (dms['d'] + dms['m'].div(60) + dms['s'].div(3600))

#Converting longitude to decimal
temp['Longitude'] = temp['Longitude'].str.replace(',', '.', regex=False)
temp['Longitude'] = temp['Longitude'].str.replace('W ', '', regex=False)
temp['Longitude'] = temp['Longitude'] + ' W'
temp['Longitude'] = temp['Longitude'].str.replace(' ', '', regex=False)

dms = temp['Longitude'].str.extract(pattern)
dms['h'] = dms['h'].map(hemisphere_sig)
dms = dms.astype(float)

temp['Longitude'] = dms['h'] * (dms['d'] + dms['m'].div(60) + dms['s'].div(3600))

temp_geo = gpd.GeoDataFrame(temp,geometry = gpd.points_from_xy(temp.Longitude,temp.Latitude))
world_data = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
axis = world_data[world_data.continent == 'South America'].plot(color = 'black',edgecolor = 'white')
temp_geo.plot(ax = axis, color = 'yellow', markersize = 20, marker='*')
plt.title('South America', fontsize = 16)
plt.xlabel('Longitude', fontsize = 16)
plt.ylabel('Latitude', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

fig = plt.gcf()
fig.set_size_inches(35,21)
#fig.savefig('matplot.png', dpi = 200)
plt.show()
```
