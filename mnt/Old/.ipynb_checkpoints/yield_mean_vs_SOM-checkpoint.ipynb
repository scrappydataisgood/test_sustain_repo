```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_winter = pd.read_csv('/mnt/Datasets/df_3_27_23_winter_parent_compounds.csv').drop(columns = ['Unnamed: 0'])
print(display(df_winter))
```

```python
median_crop_protection = df_winter['Crop_Protection_Application_Doses'].median()
high_crop_protection = df_winter['Crop_Protection_Application_Doses'].quantile(.95)
low_crop_protection = df_winter['Crop_Protection_Application_Doses'].quantile(.05)
median_N_Urea = df_winter['N_as_Urea_kg_per_Ha'].median()
high_N_Urea = df_winter['N_as_Urea_kg_per_Ha'].quantile(.95)
low_N_Urea = df_winter['N_as_Urea_kg_per_Ha'].quantile(.05)
median_N_Nitrate = df_winter['N_as_Nitrate_kg_per_Ha'].median()
high_N_Nitrate = df_winter['N_as_Nitrate_kg_per_Ha'].quantile(.95)
low_N_Nitrate = df_winter['N_as_Nitrate_kg_per_Ha'].quantile(.05)
median_N_Ammonium = df_winter['N_as_Ammonium_kg_per_Ha'].median()
high_N_Ammonium = df_winter['N_as_Ammonium_kg_per_Ha'].quantile(.95)
low_N_Ammonium = df_winter['N_as_Ammonium_kg_per_Ha'].quantile(.05)
median_P_Phosphorus = df_winter['P_as_Phosphorus_kg_per_Ha'].median()
high_P_Phosphorus = df_winter['P_as_Phosphorus_kg_per_Ha'].quantile(.95)
low_P_Phosphorus = df_winter['P_as_Phosphorus_kg_per_Ha'].quantile(.05)
median_P_P2O5 = df_winter['P_as_P2O5_kg_per_Ha'].median()
high_P_P2O5 = df_winter['P_as_P2O5_kg_per_Ha'].quantile(.95)
low_P_P2O5 = df_winter['P_as_P2O5_kg_per_Ha'].quantile(.05)
median_K_Potassium = df_winter['K_as_Potassium_kg_per_Ha'].median()
high_K_Potassium = df_winter['K_as_Potassium_kg_per_Ha'].quantile(.95)
low_K_Potassium = df_winter['K_as_Potassium_kg_per_Ha'].quantile(.05)
median_K_K2O = df_winter['K_as_K2O_kg_per_Ha'].median()
high_K_K2O = df_winter['K_as_K2O_kg_per_Ha'].quantile(.95)
low_K_K2O = df_winter['K_as_K2O_kg_per_Ha'].quantile(.05)

df_cut = df_winter[(df_winter['Crop_Protection_Application_Doses'] < high_crop_protection) & (df_winter['Crop_Protection_Application_Doses'] > low_crop_protection) & \
                  (df_winter['N_as_Urea_kg_per_Ha'] < high_N_Urea) & (df_winter['N_as_Urea_kg_per_Ha'] > low_N_Urea) & \
                  (df_winter['N_as_Nitrate_kg_per_Ha'] < high_N_Nitrate) & (df_winter['N_as_Nitrate_kg_per_Ha'] > low_N_Nitrate) &\
                  (df_winter['N_as_Ammonium_kg_per_Ha'] < high_N_Ammonium) & (df_winter['N_as_Ammonium_kg_per_Ha'] > low_N_Ammonium) & \
                  (df_winter['P_as_Phosphorus_kg_per_Ha'] < high_P_Phosphorus) & (df_winter['P_as_Phosphorus_kg_per_Ha'] > low_P_Phosphorus) & \
                  (df_winter['P_as_P2O5_kg_per_Ha'] < high_P_P2O5) & (df_winter['P_as_P2O5_kg_per_Ha'] > low_P_P2O5) & \
                  (df_winter['K_as_Potassium_kg_per_Ha'] < high_K_Potassium) & (df_winter['K_as_Potassium_kg_per_Ha'] > low_K_Potassium) &\
                  (df_winter['K_as_K2O_kg_per_Ha'] < high_K_K2O) & (df_winter['K_as_K2O_kg_per_Ha'] > low_K_K2O)]

print(display(df_cut))
```
