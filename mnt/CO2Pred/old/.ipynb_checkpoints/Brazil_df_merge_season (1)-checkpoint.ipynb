```python
import pandas as pd
import numpy as np
import datetime as dt

import os

cwd=os.path.abspath('')
os.chdir('/mnt')
files=os.listdir(cwd)
```

```python
df_1 = pd.read_excel('/mnt/Datasets/PC003_consolidated_new (1).xlsx')
df_1['Field_Name'] = df_1['Contrato legal'].str[6:]

print(display(df_1))
```

```python
pd.set_option('display.max_columns', None)
print(df_1['Field_Name'].value_counts())

print(display(df_1[df_1['Field_Name'] == 'M2V1630']))
```

```python
df_2 = pd.read_csv('/mnt/Datasets/merged_df_1_27_23_w_crop_area.csv').drop(columns = 'Unnamed: 0')
df_2 = df_2.rename(columns = {'Enter Crop Area (ha)': 'Crop_Area_all_subfields', 'Enter Dry Finished Yield Mass (kg)': 'Dry_Yield_Mass_all_subfields'})
print(display(df_2))
```

```python
df_1 = df_1[df_1['Field_Name'].isin(df_2['Field_Name'])].reset_index(drop = True)

df_1 = df_1.rename(columns = {'Safra': 'Season'})

df_1 = df_1.merge(df_2, how = 'left', on = 'Field_Name')

df_builder = pd.DataFrame(columns = df_1.columns)

unique_field_names = df_1['Field_Name'].unique()

for i in range(len(unique_field_names)):
    temp_df1 = df_1[df_1['Field_Name'] == unique_field_names[i]].reset_index(drop = True)
    
    if (len(temp_df1) > 1):
        for j in range(len(temp_df1)):
            temp_df1.loc[j, 'Field_Name'] = temp_df1.loc[j, 'Field_Name'] + '_' + str(j)
    
    df_builder = pd.concat([df_builder, temp_df1]).reset_index(drop = True)

print(display(df_builder))

df_builder.to_csv('/mnt/Datasets/merged_df_2_22_23.csv')

#when there are multiple subfields, they are labeled coded with '_x'
```
