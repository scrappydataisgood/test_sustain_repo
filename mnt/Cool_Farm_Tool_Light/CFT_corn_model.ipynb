```python
#Total Emissions in kgCO2e (kg of CO2 emissions)

def total_CO2_emissions(L_residue, L_fert_prod, L_fert, L_pest, L_fuel_and_energy, L_c_stocks, L_trans):
    return L_residue + L_fert + L_pest + L_fuel_and_energy + L_c_stocks + L_trans
```

```python
#L_pest (Pesticide Emissions)

def pesticide_emissions():
    
    output_pesticide_CO2_values = list()
    output_field_name_values = list()
    output_scenario_ID_values = list()
    
    field_data = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn_input.xlsx', \
                               sheet_name = 'field_characteristics')
    data = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn_input.xlsx', \
                         sheet_name = 'L_pest')
    table = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn.xlsx', sheet_name = 'EF_pest')
    
    field_names = field_data['field_name'].unique()
    
    print(display(data))
    print(display(table))
    
    for i in range(len(field_names)):
        
        data_temp = data[data['field_name'] == field_names[i]].reset_index(drop = True)
        field_data_temp = field_data[field_data['field_name'] == field_names[i]].reset_index(drop = True)
        
        scenario_IDs_temp = data_temp['scenario_ID'].unique()
        
        for j in range(len(scenario_IDs_temp)):
            
            data_temp2 = data_temp[data_temp['scenario_ID'] == scenario_IDs_temp[j]].reset_index(drop = True)
            
            L_pest = 0
            
            for k in range(len(data_temp2)):
            
                area_ha = field_data_temp['area_ha'].values[0]
                Rate_app_kg_per_ha = data_temp2['Rate_app_kg_per_ha'].values[k]
                AI_frac = data_temp2['AI_frac'].values[k]
                EF_pest = table[table['pesticide_type'] == \
                                data_temp2['pesticide_type'].values[k]]['climate_change'].values[0]
                
                L_pest += area_ha*Rate_app_kg_per_ha*AI_frac*EF_pest
            
            output_field_name_values.append(field_names[i])
            output_scenario_ID_values.append(scenario_IDs_temp[j])
            output_pesticide_CO2_values.append(L_pest)
        
    return pd.DataFrame({'Field Name': output_field_name_values, 'Scenario ID': output_scenario_ID_values, \
                         'CO2_emissions': output_pesticide_CO2_values})

print(display(pesticide_emissions()))
```

```python
#L_Transport (Transport Emissions)

def transport_emissions():
    
    output_transport_CO2_values = list()
    output_scenario_ID_values = list()
    output_field_name_values = list()
    
    field_data = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn_input.xlsx', \
                               sheet_name = 'field_characteristics')
    data = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn_input.xlsx', \
                         sheet_name = 'L_transport')
    table = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_light_corn.xlsx', sheet_name = 'EF_VEH')
    
    field_names = field_data['field_name'].unique()
    
    print(display(data))
    print(display(table))
    for i in range(len(field_names)):
        
        data_temp = data[data['field_name'] == field_names[i]].reset_index(drop = True)
        field_data_temp = field_data[field_data['field_name'] == field_names[i]].reset_index(drop = True)
        
        scenario_IDs_temp = data_temp['scenario_ID'].unique()
        
        for j in range(len(scenario_IDs_temp)):  
            
            data_temp2 = data_temp[data_temp['scenario_ID'] == scenario_IDs_temp[j]].reset_index(drop = True) 
            
            L_transport = 0
            
            for k in range(len(data_temp2)):
                
                EF_VEH = table[table['transportation_mode'] == \
                               data_temp2['transportation_mode'].values[k]]['impact'].values[0]
                D = data_temp2['distance_km'].values[k]
                M = data_temp2['transported_mass_tonnes'].values[k]
                L_transport += EF_VEH*D*M
            
            output_field_name_values.append(field_names[i])
            output_scenario_ID_values.append(scenario_IDs_temp[j])
            output_transport_CO2_values.append(L_transport)
                
    return pd.DataFrame({'Field Name': output_field_name_values, 'Scenario ID': output_scenario_ID_values, \
                         'CO2_emissions': output_transport_CO2_values})

print(display(transport_emissions()))
```

```python
#L_fert (Fertilizer Emissions)

#all in same exponent for N2O and NO, while separate exponents for NH3?

def fertilizer_emissions():
    
    fertilizer_CO2_values = list()
    
    data = pd.read_excel('CFT_light_corn_input.xlsx', sheet_name = 'L_fert_prod')
    table = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'EF_fert')
    table2 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'EF_compost')
    table3 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'EF_custom_fert_prod')
    
    table4 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_constant')
    table5 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_fert_type')
    table6 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_soil_texture')
    table7 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_CEC')
    table8 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_SOC')
    table9 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_pH')
    table10 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_drainage')
    table11 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_moisture')
    table12 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'Bouwman_application_method')
    table13 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'bulk_density')
    table14 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'clay_content')
    table15 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'carbon_density')
    
    field_names = data['field_name'].unique()
    
    for i in range(len(field_names)):
        
        data_temp = data[data['field_name'] == field_names[i]].reset_index(drop = True)
        
        scenario_IDs_temp = data_temp['scenario_ID'].unique()
        
        for j in range(len(scenario_IDs_temp)):
            
            data_temp2 = data_temp[data_temp['scenario_ID'] == scenario_IDs_temp[j]].reset_index(drop = True)
            
            L_fert_prod = 0
            L_direct_N2O = 0
            L_back_N2O = 0
            L_direct_NO = 0
            L_back_NO = 0
            L_vol = 0
            L_leach = 0
            L_limestone = 0
            L_urea = 0
            
            #Fertilizer induced N2O emissions#
            f_k_N2O = table4['N2O'].values[0]
            f_soil_texture_N2O = table6[table6['soil_texture'] == data_temp2['soil_texture'].values[k]]['N2O'].values[0]
            f_SOC_N2O = table8[(table8['SOC_UB'] >= data_temp2['SOC'].values[k]) & \
                              (table8['SOC_LB'] < data_temp2['SOC'].values[k])]['N2O'].values[0]
            f_soil_pH_N2O = table9[(table9['pH_UB'] >= data_temp2['pH'].values[k]) & \
                              (table9['pH_LB'] < data_temp2['pH'].values[k])]['N2O'].values[0]
            f_drainage_N2O = table10[table10['drainage'] == data_temp2['drainage'].values[k]]['N2O'].values[0]
            f_soil_moisture_N2O = table11[table11['moisture'] == data_temp2['moisture'].values[k]]['N2O'].values[0]
            L_direct_N2O_exp = f_k_N2O + f_fert_type_N2O + f_soil_texture_N2O + f_SOC_N2O + f_soil_pH_N2O + f_drainage_N2O + f_soil_moisture_N2O
            L_back_N2O_exp = f_k_N2O + f_soil_texture_N2O + f_SOC_N2O + f_soil_pH_N2O + f_drainage_N2O + f_soil_moisture_N2O
            
            #Fertilizer induced NO emissions#
            f_k_NO = table4['NO'].values[0]
            f_fert_type_NO = table5[table5['fertilizer_type'] == data_temp2['fertilizer_or_compost_type'].values[k]]['NO'].values[0]
            f_SOC_NO = table8[(table8['SOC_UB'] >= data_temp2['SOC'].values[k]) & \
                              (table8['SOC_LB'] < data_temp2['SOC'].values[k])]['NO'].values[0]
            f_drainage_NO = table10[table10['drainage'] == data_temp2['drainage'].values[k]]['NO'].values[0]
            f_soil_moisture_NO = table11[table11['moisture'] == data_temp2['moisture'].values[k]]['NO'].values[0]
            L_direct_NO_exp = f_k_NO + f_fert_type_NO + f_SOC_NO + f_drainage_NO
            L_back_NO_exp = f_k_NO + f_SOC_NO + f_drainage_NO
            
            for k in range(len(data_temp2)):
                
                application_rate = data_temp2['application_rate_kg'].values[k]
                
                ##############emissions from fertilizer production
                
                if (data_temp2['fertilizer_type'].values[k] == 'Compose-your-own'):
                    
                    K2O_contribution = data_temp2['K2O_proportion'].values[k]*table3[table3['geographic_location'] == data_temp2['geographic_location'].values[k]]['K2O'].values[0]
                    ammonia_N_contribution = data_temp2['ammonia_N_proportion'].values[k]*table3[table3['geographic_location'] == data_temp2['geographic_location'].values[k]]['ammonia_N'].values[0]
                    nitric_acid_N_contribution = data_temp2['nitric_acid_N_proportion'].values[k]*table3[table3['geographic_location'] == data_temp2['geographic_location'].values[k]]['nitric_acid_N'].values[0]
                    urea_N_contribution = data_temp2['urea_N_proportion'].values[k]*table3[table3['geographic_location'] == data_temp2['geographic_location'].values[k]]['urea_N'].values[0]
                    P2O5_contribution = data_temp2['P2O5_proportion'].values[k]*table3[table3['geographic_location'] == data_temp2['geographic_location'].values[k]]['P2O5'].values[0]
                    EF_fert = K2O_contribution + ammonia_N_contribution + nitric_acid_N_contribution + urea_N_contribution + P2O5_contribution
                    
                elif ((data_temp2['fertilizer_type'].values[k] == 'Compost (zero emissions)') or \
                     (data_temp2['fertilizer_type'].values[k] == 'Compost (fully aerated production)') or \
                     (data_temp2['fertilizer_type'].values[k] == 'Compost (non-fully aerated production)')):
                    
                    EF_fert = table2[table2['compost_type'] == data_temp2['fertilizer_or_compost_type'].values[k]]['EF'].values[0]
                    
                else:
                    EF_fert = table2[(table2['fertilizer_type'] == data_temp2['fertilizer_or_compost_type'].values[k]) & \
                                    (table2['geographic_location_and_year'] == data_temp2['geographic_location'].values[k])]['emissions'].values[0]
            
                L_fert_prod += application_rate*EF_fert
            
                ##############soil fertilizer induced
                
                GWP_N2O = 273
            
                #N2O
                
                f_fert_type_N2O = table5[table5['fertilizer_type'] == data_temp2['fertilizer_or_compost_type'].values[k]]['N2O'].values[0]*application_rate
                L_direct_N2O_exp += f_fert_type_N2O
    
                #NO
        
                f_fert_type_NO = table5[table5['fertilizer_type'] == data_temp2['fertilizer_or_compost_type'].values[k]]['NO'].values[0]*application_rate
                L_direct_NO_exp += f_fert_type_NO
                
                #NH3
                
                f_fert_type_vol = table5[table5['fertilizer_type'] == data_temp2['fertilizer_or_compost_type'].values[k]]['NH3'].values[0]
                f_soil_pH_vol = table9[(table9['pH_UB'] >= data_temp2['pH'].values[k]) & \
                                  (table9['pH_LB'] < data_temp2['pH'].values[k])]['NH3'].values[0]
                bulk_density = table13[data_temp2['soil_texture'].values[k]].values[0]
                clay_content = table14[data_temp2['soil_texture'].values[k]].values[0]
                carbon_density = table15[(table15['SOM_LB'] < data_temp2['SOM'].values[k]) & (table15['SOM_UB'] >= data_temp2['SOM'].values[k])]['carbon_density'].values[0]
                SOC_t = bulk_density*carbon_density*1000
                CEC = (51*data_temp2['pH'].values[k] - 59)*SOC_t/(3000000*bulk_density) + (30 + 4.4*data_temp2['pH'].values[k])*clay_content
                
                f_soil_moisture_vol = table11[table11['moisture'] == data_temp2['moisture'].values[k]]['NH3'].values[0]    
                f_application_method_vol = table12[table12['application_method'] == data_temp2['application_method'].values[k]]['NH3'].values[0] 
                L_vol += .01*application_rate*np.exp(f_fert_type_vol + f_soil_pH_vol + f_soil_moisture_vol + f_application_method_vol)
                
                #Leach#
                
                if (data['moisture'].values[k] == 'moist'):
                    F_leach = .3
                    EF_leach = .0075
                    L_Leach += application_rate*F_leach*EF_leach
                
                #Limestone# double check***
                
                EF_limestone = .12
                EF_dolomite = .13
                
                if (data_temp2['fertilizer_or_compost_type'].values[k] == 'Limestone - 55 % CaCO3 / 29 % CaO'):
                    L_limestone += application_rate*EF_limestone*44/12
                if (data_temp2['fertilizer_or_compost_type'].values[k] == 'Dolomite - CaMg(CO3)2'):
                    L_limestone += application_rate*EF_dolomite*44/12

                #Urea
                
                if ((data_temp2['fertilizer_or_compost_type'].values[k] == 'Urea - 46 % N') or \
                   (data_temp2['fertilizer_or_compost_type'].values[k] == 'Urea - 46.4 % N') or \
                   (data_temp2['fertilizer_or_compost_type'].values[k] == 'Urea ammonium nitrate solution - 32 % N')):
                    EF_urea = .2
                    L_urea += EF_urea*application_rate*44/12
            
            L_direct_N2O = np.exp(L_direct_N2O_exp)
            L_back_N2O = np.exp(L_back_N2O_exp)
            L_direct_NO = .01*np.exp(L_direct_NO_exp)
            L_back_NO = .01*np.exp(L_back_NO_exp)
            
            if (data_temp2['nitrification_inhibitor_binary'].values[0] == True): ##double check this
                L_direct_N2O = .63*L_direct_N2O
                L_direct_NO = .54*L_direct_NO

            L_fert = L_fert_prod + L_direct_N2O + L_back_N2O + L_direct_NO + L_back_NO + L_vol + L_leach + L_limestone + L_urea
            
            fertilizer_CO2_values.append(L_fert)
    
    return fertilizer_CO2_values
```

```python
#Fuel and Energy Use Emissions

def fuel_and_energy_emissions():
    
    fuel_and_energy_CO2_values = list()
    
    data = pd.read_excel('CFT_light_corn_input.xlsx', sheet_name = 'L_fuel_and_energy')
    table = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'EF_fuel_and_energy')
    table2 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'FR_machinery')
    table3 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'ER')
    table4 = pd.read_excel('CFT_light_corn.xlsx', sheet_name = 'EF_irri_location')
    
    field_names = data['field_name'].unique()
    
    for i in range(len(field_names)):
        
        data_temp = data[data['field_name'] == field_names[i]].reset_index(drop = True)
        
        scenario_IDs_temp = data_temp['scenario_ID'].unique()
        
        for j in range(len(scenario_IDs_temp)):
            
            data_temp2 = data_temp[data_temp['scenario_ID'] == scenario_IDs_temp[j]].reset_index(drop = True)
            
            L_fuel_and_energy = 0
            
            for k in range(len(data_temp2)):
                
                #direct energy use
                if (~data_temp2['fuel_amount'].values[k].isna()):
                    
                    C_fuel_and_energy = data_temp2['fuel_amount'].values[k]
                    EF_fuel_and_energy = table[table['energy_name'] == data_temp2['energy_name'].values[k]]
                    L_fuel_and_energy += C_fuel_and_energy*EF_fuel_and_energy
                
                
                elif (~data_temp2['machinery_name'].values[k].isna()):
                    
                    #field operations (harvest)
                    if (data_temp2['machinery_name'].values[k] == 'harvesting'):
                        P = data_temp2['power_requirements'].values[k]
                        a = 2.64
                        c = 3.91
                        d = 738
                        e = 173

                        if (data_temp2['power_requirements'].values[k] <= 75):
                            b = np.max(.2, P/75)
                        else:
                            b = 1

                        F = a*b + c - .203*(d*b + e)^(.5)
                        FR_machinery = F*P
                        
                    #field operations (tillage and sparying / spreading)
                    else:     
                        FR_machinery = table2[(table2['FR_machinery'] == data_temp2['machinery_type'].values[k])][data_temp2['soil_texture'].values[k]].values[0]
                        area_ha = data_temp2['area_ha'].values[k]
                        
                        #number of field operations
                        if(data_temp2['machinery_name'].values[k].str.contains(['bale', 'Bale'], na=False, regex=True)):
                            num_operations = data_temp2['num_bales'].values[k]
                            L_fuel_and_energy += num_operations*FR_machinery*area_ha
                        
                        #bales of hay
                        else:
                            num_bales = data_temp2['num_operations'].values[k]
                            L_fuel_and_energy += num_bales*FR_machinery*area_ha
                    
                #irrigation energy use  -- may want to break up irrigation and other energy
                else:
                    D_irri = data_temp2['irrigation_depth_mm'].values[k]
                    A_irri = data_temp2['irrigation_area_ha'].values[k]
                    
                    #by values
                    if((~data_temp2['l_horizontal_m'].values[k].isna()) and (~data_temp2['l_vertical_m'].values[k].isna())):
                        
                        ER_irri_method = table3[table3['irrigation_method'] == data_temp2['irrigation_method'].values[k]][data_temp2['irrigation_fuel_type'].values[k]].values[0]
                        l_horizontal_m = data_temp2['l_horizontal_m'].values[k] 
                        l_vertical_m = data_temp2['l_vertical_m'].values[k]
                        ER_h_trans = .95
                        ER_v_trans = .095
                        ER_irri_area = ER_irri_method + l_horizontal_m*ER_h_trans + l_vertical_m*ER_v_trans
                        
                    #by region (double check)
                    else:
                        ER_irri = table4[(table4['method'] == data_temp2['irrigation_method'].values[k]) and \
                                        (table4['power_source'] == data_temp2['irrigation_power_source'].values[k]) and \
                                        (table4['region'] == data_temp2['region_code'].values[k])]['emission_factor'].values[0]
                        
                    L_fuel_and_energy += ER_irri_area*D_irri*A_irri
                    
            fuel_and_energy_CO2_values.append(L_fuel_and_energy)
    
    return fuel_and_energy_CO2_values
```
