```python
def swap_rows(df, row1, row2):
    df.iloc[row1], df.iloc[row2] =  df.iloc[row2].copy(), df.iloc[row1].copy()
    return df

fertilizer_CO2_values = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_fertilizer_CO2_values.xlsx')

print(display(fertilizer_CO2_values))

for k in range(8):
    ratio_values_general = list()
    fertilizer_names = fertilizer_CO2_values['fertilizer'].unique()

    for j in range(len(fertilizer_names) - 1):

        ratio_values = list()

        for i in range(len(fertilizer_CO2_values.columns) - 1):
            fertilizer_rate_1 = fertilizer_CO2_values[fertilizer_CO2_values.columns[i + 1]][j+1].astype(np.float32)
            fertilizer_rate_2 = fertilizer_CO2_values[fertilizer_CO2_values.columns[i + 1]][0].astype(np.float32)
            ratio_values.append(fertilizer_rate_1/fertilizer_rate_2)

        ratio_values_general.append(ratio_values)

    app_rates = [100,125,150,175,200,225,250]

    plt.figure(figsize = (8,8))
    for i in range(7):
        plt.plot(app_rates,ratio_values_general[i])
        plt.title('CO2 Emissions Ratio between Different Fertilizers and ' + fertilizer_CO2_values['fertilizer'].unique()[0], fontsize = 14)
        plt.xlabel('Application Rate (kg per Ha)', fontsize = 14)
        plt.ylabel('Ratio', fontsize = 14)
        plt.legend(fertilizer_names[1:], bbox_to_anchor = [1, -.1], fontsize = 14)
        plt.xticks(fontsize = 14)
        plt.yticks(fontsize = 14)
    plt.show()
        
    fertilizer_CO2_values = pd.read_excel('/mnt/Cool_Farm_Tool_Light/CFT_fertilizer_CO2_values.xlsx')
    if (k == 7):
        break
    fertilizer_CO2_values = swap_rows(fertilizer_CO2_values, 0, k+1)

# print(display(df))
    
    

        
        
        
```
